```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams['text.usetex'] = True
plt.rcParams['font.family'] = 'serif'
plt.rcParams['font.serif'] = ['Computer Modern Roman']
plt.rcParams.update({'font.size': 16, 'axes.labelsize': 16, 'axes.titlesize': 16, 'legend.fontsize': 14})

# Data for the plot
datasets = ['lineitem', 'acs', 'iot', 'icu']
approach_1_times = np.array([37.1, 6.4, 13.6, 11.6])
approach_2_times = np.array([46.8, 9.1, 12.3, 10.3])
approach_3_times = np.array([149.4, 55.8, 57.2, 38.2])
approach_4_times = np.array([358.6, 24.5 , 48.7, 45.8])
approach_5_times = np.array([392, 86.8, 164.3, 90.1])

# Names and colors for the approaches
approaches = ['xdbc', 'connector-x','duckdb', 'modin[ray]', 'turbodbc']

# Formal, colorblind-friendly palette
formal_palette = sns.color_palette("colorblind", 5)
colors = formal_palette  # Using formal palette

# Bar width and positions for the groups
bar_width = 0.17
x_indexes = np.arange(len(datasets))

# Create the plot
plt.figure(figsize=(6, 3.75))

# Plotting each approach with offset for bar positions
plt.bar(x_indexes - 2 * bar_width, approach_1_times, width=bar_width, color=colors[0], label=approaches[0], zorder=3)
plt.bar(x_indexes - bar_width, approach_2_times, width=bar_width, color=colors[1], label=approaches[1], zorder=3)
plt.bar(x_indexes, approach_3_times, width=bar_width, color=colors[2], label=approaches[2], zorder=3)
plt.bar(x_indexes + bar_width, approach_4_times, width=bar_width, color=colors[3], label=approaches[3], zorder=3)
plt.bar(x_indexes + 2 * bar_width, approach_5_times, width=bar_width, color=colors[4], label=approaches[4], zorder=3)



# Labels and Title
plt.xlabel('Datasets')
plt.ylabel('Time (s)')
#plt.title('Comparison of Approaches on Different Datasets', fontsize=14)

# X-axis ticks and legend
plt.xticks(ticks=x_indexes, labels=datasets)
plt.legend(loc='best')

# Grid for better readability
plt.grid(axis='y', alpha=0.3, zorder=0)

# Display the plot
plt.tight_layout()
#plt.show()
plt.savefig(f'../opt_paper_plots/pandas_postgres.pdf', bbox_inches='tight')
```
